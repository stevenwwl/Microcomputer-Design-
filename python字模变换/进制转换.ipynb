```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('zimo.txt', header=None)
```

```python
df
```

```python
new_df = df.copy()
```

```python
for i in range(16):
    for j in range(1,17):
        new_df.iloc[i,j] = '{:08b}'.format(int(df.iloc[i,j],16))
```

```python
new_df
```

```python
for i in range(16):
    for j in range(1,17):
        new_df.iloc[i,j] = ''.join(reversed(new_df.iloc[i,j]))
```

```python
new_df
```

```python
for i in range(16):
    for j in range(1,17):
        new_df.iloc[i,j] = hex(int(new_df.iloc[i,j],2))
```

```python
new_df
```

```python
for i in range(16):
    for j in range(1,17):
        new_df.iloc[i,j] = new_df.iloc[i,j][2:]
```

```python
new_df
```

```python
for i in range(16):
    for j in range(1,17):
        new_df.iloc[i,j] = '{:0>2}'.format(new_df.iloc[i,j])
```

```python
new_df
```

```python
for i in range(16):
    for j in range(1,17):
        new_df.iloc[i,j] = new_df.iloc[i,j].upper()
```

```python
new_df
```

```python
for i in range(16):
    for j in range(1,17):
        new_df.iloc[i,j] = new_df.iloc[i,j]+'H'
```

```python
new_df
```

```python
new_df.iloc[:,0] = 'DB '+ new_df.iloc[:,1]
```

```python
new_df
```

```python
new_df.drop(columns=1, inplace=True)
```

```python
new_df
```

```python
new_df.to_csv('right_zimo.txt', header=False, index=False)
```
